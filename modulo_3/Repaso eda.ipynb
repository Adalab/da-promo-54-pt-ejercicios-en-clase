"Int 64" Es un tipo especial de dato en pandas:

- Int64Dtype no existe en NumPy; es una extensión de pandas.
- Está diseñado para permitir enteros y valores nulos (\<NA>) sin convertir a float64.

| Tipo      | Soporta NaN | Representación de NaN | Conversión implícita            |
| --------- | ----------- | --------------------- | ------------------------------- |
| int64   | ❌           | No permitido          | Error si hay NaN                |
| float64 | ✔           | np.nan              | Sí, pero pierde enteros exactos |
| Int64   | ✔           | <NA> (pandas NA)    | No convierte a float            |

```python
# iterative imputer y knn imputer 

columnas_avanzada = ['age', 'euribor3m']

iterative_imputer = IterativeImputer(max_iter=20, random_state=42)
iterative_imputado = iterative_imputer.fit_transform(df[columnas_avanzada])
df[['age_iterative', 'euribor3m_iterative']] = iterative_imputado
```
